
1. **Install Python and pip**
   If you haven't already, you'll need to install Python and pip. You can download Python from the [official website](https://www.python.org/downloads/). Pip is included in Python 3.4 and later versions.

2. **Install Jupyter Notebook/Lab**
   You can install Jupyter Notebook or Jupyter Lab using pip. Open your terminal and type the following commands:

   ```python
   # For Jupyter Notebook
   pip install notebook

   # For Jupyter Lab
   pip install jupyterlab
   ```

3. **Launch Jupyter Notebook/Lab**
   After installation, you can launch Jupyter Notebook or Jupyter Lab by typing the following commands in your terminal:

   ```python
   # For Jupyter Notebook
   jupyter notebook

   # For Jupyter Lab
   jupyter lab
   ```

   This will open a new tab in your default web browser that shows the Notebook Dashboard.

4. **Create a new notebook**
   In the Notebook Dashboard, you can create a new notebook by clicking on the "New" button and selecting "Python 3" (or the version you have installed).

5. **Use the notebook**
   You can now start using the notebook. You can type Python code in the cells, and then run the code by clicking the "Run" button or using the keyboard shortcut Shift+Enter.

6. **Save and exit**
   You can save your notebook by clicking the "Save" button or using the keyboard shortcut Ctrl+S. To exit the notebook, you can simply close the browser tab and then stop the server in your terminal by pressing Ctrl+C.

Remember to replace `pip` with `pip3` and `python` with `python3` if you're using Python 3. If you're on a system with both Python 2 and Python 3, you might need to use `pip3` and `python3` instead of `pip` and `python`.
